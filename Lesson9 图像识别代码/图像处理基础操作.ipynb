```python
import os
filename = './mydata/rice dumpling.jpg'
```

```python
#matplotlib
import matplotlib.pyplot as plt

img=plt.imread(filename)             # plt.imread()-----np.ndarray,  (H x W x C), [0, 255],RGB
print(img.shape)
print(img[1,1,:])   #rgb
plt.imshow(img)
plt.show()
```

```python
modify_img=img.copy()
modify_img[:,:,0]=0
modify_img[:,:,1]=0
#modify_img[:,:,2]=0
plt.imshow(modify_img)
plt.show()
```

```python
change_img = img.copy()
change_img[100:600,100:600,:]=[0,255,0]
plt.imshow(change_img)
plt.show()
```

```python
# PIL
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img_pil = Image.open(filename)         # PIL.Image.Image
img_pil_1 = np.array(img_pil)           # (H x W x C), [0, 255], RGB
print(img_pil_1.shape)
print(img_pil_1[1,1,:])
plt.imshow(img_pil_1)                   #plt.imshow(img_pil)
plt.show()
```

```python
import torchvision.transforms as T
my_transform=T.Compose([
    T.CenterCrop(size=(400,400)), #400
#    T.RandomCrop(size=(400,400)), # 400
    T.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5, hue=0),   #
    T.RandomHorizontalFlip(p=0.5), # 0.5
    T.RandomVerticalFlip(p=0.5), # 0.5
#   T.RandomRotation(degrees=5, expand=False, fill=False),
    T.ToTensor(),                                            #ndarraytensor(H x W x C)C x H x W,int[0, 255]float[0, 1]
#    T.Normalize([0.4385,0.4422,0.4460],[0.183,0.174,0.182])  #
])
img_t=my_transform(img_pil)
#print(img_t.shape)
#print(img_t[:,1,1])

#
img_t_np=img_t.numpy().transpose(1, 2, 0)
print(img_t_np.shape)
plt.imshow(img_t_np)
plt.show()
```
