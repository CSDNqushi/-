```python
import os
filename = './mydata/rice dumpling.jpg'
```

```python
#使用matplotlib读取
import matplotlib.pyplot as plt

img=plt.imread(filename)             # plt.imread()-----np.ndarray,  (H x W x C), [0, 255],RGB
print(img.shape)
print(img[1,1,:])   #显示某个像素点的rgb值
plt.imshow(img)
plt.show()
```

```python
modify_img=img.copy()
modify_img[:,:,0]=0
modify_img[:,:,1]=0
#modify_img[:,:,2]=0
plt.imshow(modify_img)
plt.show()
```

```python
change_img = img.copy()
change_img[100:600,100:600,:]=[0,255,0]
plt.imshow(change_img)
plt.show()
```

```python
# 使用PIL读取
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img_pil = Image.open(filename)         # PIL.Image.Image对象
img_pil_1 = np.array(img_pil)           # (H x W x C), [0, 255], RGB
print(img_pil_1.shape)
print(img_pil_1[1,1,:])
plt.imshow(img_pil_1)                   #也可直接plt.imshow(img_pil)
plt.show()
```

```python
import torchvision.transforms as T
my_transform=T.Compose([
    T.CenterCrop(size=(400,400)), #将图像中心裁剪，宽高均为400
#    T.RandomCrop(size=(400,400)), # 将图像随意裁剪，宽高均为400
    T.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5, hue=0),   #调整图像亮度、对比度等
    T.RandomHorizontalFlip(p=0.5), # 以0.5的概率左右翻转图像
    T.RandomVerticalFlip(p=0.5), # 以0.5的概率左右翻转图像
#   T.RandomRotation(degrees=5, expand=False, fill=False),
    T.ToTensor(),                                            #将ndarray转为tensor，从(H x W x C)转为（C x H x W）,从int类型[0, 255]转为float类型[0, 1]
#    T.Normalize([0.4385,0.4422,0.4460],[0.183,0.174,0.182])  #像素值标准化
])
img_t=my_transform(img_pil)
#print(img_t.shape)
#print(img_t[:,1,1])

#显示图像
img_t_np=img_t.numpy().transpose(1, 2, 0)
print(img_t_np.shape)
plt.imshow(img_t_np)
plt.show()
```
